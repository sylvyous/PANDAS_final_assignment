Binning Prices

```python
adf=df.copy()
bins = [0, 100000, 200000, 300000, 400000, 500000, 600000, 700000, 800000]
labels = ['0-100k', '100k-200k', '200k-300k', '300k-400k', '400k-500k', '500k-600k', '600k-700k', '700k-800k']

adf['bin_price'] = pd.cut(adf['price'], bins=bins, labels=labels, right=False)
adf

```

```python
perc=adf.groupby('bin_price')['id'].count().to_frame().reset_index()
perc['percentage'] = perc['id'].apply(lambda x: 100 * x / 18445)
perc
```

Around 45 percent of the houses fall between the 300k to 500k bracket

Grade

```python
grd=adf.copy()
```

```python
grd['cat_grade'] = grd['grade'].apply(lambda x: 'very_high' if x > 8 else 'high' if 4 <= x <= 8 else 'low' if x < 4 else 'unknown')
grd[['cat_grade','bin_price','sqft_living','sqft_living15']].groupby(['bin_price','cat_grade']).mean().round(1).reset_index()
#sqft_living15 - The square footage of interior housing living space for the nearest 15 neighbors
```

All of the houses starting from 200K are either high or very high grade
Low grade houses are on average very small in terms of living area and compared to the houses in the same neighborhood 
they are on average much much smaller.
The higher we go on the grading scale, the closer the sqft_living of the house matches that of the neighboring houses

Waterfront and View

```python
adf[['bin_price', 'waterfront', 'view', 'grade']].groupby('bin_price').agg(['max', 'min', 'mean', 'median']).reset_index()
```

```python
adf[['bin_price', 'waterfront', 'view', 'grade']].groupby(['waterfront','bin_price']).mean().round(1).reset_index()
```

This shows us that the grade and waterfront aren't very strongly correlated, They remain pretty consistent for both waterfront and not waterfront, price seems to have more of an effect on grade.
Waterfront is more closely related to the view. But nither have much of an effect on the price in our dataset


YEAR BUILT



```python
ydf=df.copy()
ydf['flag_yr_built']=adf.apply(lambda x:1 if x['yr_built']>1995 else 0, axis=1)
adf.groupby('flag_yr_built').agg({'price' : ['mean', 'max', 'min'],
                             'sqft_living' : ['mean', 'max', 'min'],
                             'sqft_lot' : ['mean', 'max', 'min'],
                             'condition' : ['median', 'max', 'min'],
                             'grade' : ['median', 'max', 'min'],
                             'id' : ['count']}).round(1)
```

```python
ydf.groupby('flag_yr_built').agg({'price' : ['mean', 'max', 'min'],
                             'bathrooms' : ['mean', 'max', 'min'],
                             'bedrooms' : ['mean', 'max', 'min'],
                             'floors' : ['median', 'max', 'min'],
                             'id' : ['count']}).round(1)
```

Newer houses built have a higher grade, They are built much larger.
while the average number of bedrooms doesn't vary too much compared to houses built prior to 1995, It seems to be that new houses 
have overall more bathrooms built in, and the units generally are multi storeyed.

```python
zip=ydf[['zipcode', 'price', 'flag_yr_built','id']].groupby('zipcode').agg({'id' : ['count']}).round(1).reset_index()
zip.loc[ydf['flag_yr_built']==1]
```

Cities 98052, 98023, 98058 have the top 3 highest number of new housings built. Indicating more demand in housing.
